Implemente el descenso de gradiente estocástico, el descenso de gradiente y el descenso de gradiente por mini batches para una función polinómica de grado 3. Luego, grafique la función aproximada por cada uno de los métodos solicitados y la distribución real de puntos. Para esto considere que:
● El polinomio que debe usar es 2 * x**3 - 3 * x**2 + 5 * x + 3


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def polynomial (x):
    return 2*x**3 - 3*x**2 +5*x+3

def gradient(x):
    return 6 * x**2 - 6 * x + 5
```

```python
# Descenso de Gradiente Estocástico (SGD)
def sgd(x, learning_rate=0.01, epochs=100):
    coefficients = np.random.rand(4)  # Inicializa coeficientes aleatorios
    for epoch in range(epochs):
        for point in x:
            gradient_value = gradient(point)
            coefficients -= learning_rate * gradient_value * np.power(point, np.arange(3, -1, -1))
    return coefficients
```

```python
# Generar datos de entrenamiento con ruido
np.random.seed(42)
x_train = np.random.uniform(-5, 5, 100)
y_train = polynomial(x_train) + np.random.normal(0, 5, size=len(x_train))  # Agregar ruido normal

# Entrenar modelo SGD
coefficients_sgd = sgd(x_train)

# Generar puntos predichos por el modelo SGD
y_pred_sgd = np.polyval(coefficients_sgd[::-1], x_train)

# Graficar resultados
plt.scatter(x_train, y_train, label='Datos reales con ruido')
plt.plot(x_train, y_pred_sgd, label='SGD', color='red')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Aproximación Polinómica con SGD')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Función polinómica de grado 3
def polynomial(x):
    return 2 * x**3 - 3 * x**2 + 5 * x + 3

# Derivada de la función polinómica
def gradient(x):
    return 6 * x**2 - 6 * x + 5

# Descenso de Gradiente Estocástico (SGD)
def sgd(x, learning_rate=0.01, epochs=100):
    coefficients = np.random.rand(4)  # Inicializa coeficientes aleatorios
    for epoch in range(epochs):
        for point in x:
            gradient_value = gradient(point)
            coefficients -= learning_rate * gradient_value * np.power(point, np.arange(3, -1, -1))
    return coefficients  # Agregamos esta línea para devolver los coeficientes

# Generar datos de entrenamiento sin ruido
np.random.seed(42)
x_train = np.random.uniform(-5, 5, 100)
y_train_real = polynomial(x_train)

# Agregar ruido a los datos reales
noise = np.random.normal(0, 5, size=len(x_train))
y_train = y_train_real + noise

# Entrenar modelo SGD
coefficients_sgd = sgd(x_train)

# Generar puntos predichos por el modelo SGD
y_pred_sgd = np.polyval(coefficients_sgd[::-1], x_train)

# Graficar resultados
plt.scatter(x_train, y_train, label='Datos reales con ruido')
plt.plot(x_train, y_pred_sgd, label='SGD', color='red')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Aproximación Polinómica con SGD')
plt.show()
```
